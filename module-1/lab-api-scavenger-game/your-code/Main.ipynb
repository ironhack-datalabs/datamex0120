```python
repo = "datamex0120"
base_url = "https://api.github.com/"

for owner in l_user:
    forks_2 = base_url+"repos/{}/{}/forks".format(owner,repo)
    response_2 = requests.get(forks_2)
    results_2 = response_2.json()
    print(owner,' ',len(results_2))
    if len(results_2) == 0:
        continue
    else:
        forks_df_2 =pd.DataFrame(json_normalize(results_2))
        forks_df = forks_df.append(forks_df_2)
```
