```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#Asignacion de token para acceso

with open('token.txt', 'r') as f:
    acceso=f.readlines()[0].split('=')

user=acceso[0].lstrip('\ufeff')
key=acceso[1].rstrip('\n')
```

```python
#Conexion y descarga de la informacion desde el API de Github

repo='ironhack-datalabs/datamex0120'
url_forks = 'http://api.github.com/repos/'+ repo +'/forks'

res_fork = requests.get(url_forks, auth=(user, key))
res=res_fork.json()

print('Status servidor:',res_fork.status_code)
```

```python
res_flat=json_normalize(res)
df_res_flat=pd.DataFrame(res_flat)
df_res_flat.columns
```

```python
#Obtain the full list of forks created from the main lab repo via Github API.
df_res_flat
```

```python
list_forks=[]
for i in range(len(df_res_flat)):
    lista_forks.append(df_res_flat['full_name'][i])


for i in range(len(df_res_flat)):
    if df_res_flat['forks'][i]>0:
        reqx = requests.get(df_res_flat['forks_url'][i])
        res_flat_temp=json_normalize(reqx.json())
        df_res_flat_temp=pd.DataFrame(res_flat_temp)
        for j in range(len(df_res_flat_temp)):
            lista_forks.append(df_res_flat_temp['full_name'][j])
        
    
```
