### 8. Calculate the percentage of "Four short links" entry titles.

```python
# obtencion de los titulos en entries y almacenamiento en lista
lst = []
for i in range(len(d.entries)):
    lst.append(d.entries[i]['title'])
# si la palabra inicia con Fo, se agrega al contador 
count = 0
for i in lst:
    if i.startswith('Fo'):
        count+=1
per = round(count/len(lst) * 100, 2)
print(per,'%')
```

### 9. Create a Pandas data frame from the feed's entries.

```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
for i in range(len(d.entries)):
    print(d.entries[i].content)
```

```python
df = pd.DataFrame(d.entries)
df.summary
```

### 10. Count the number of entries per author and sort them in descending order.

```python
js = json_normalize(d.entries)
df = pd.DataFrame(js)
gb = df.groupby('author').count()
sort = gb.sort_values(by='title',ascending=False)
sort
```

### 11. Add a new column to the data frame that contains the length (number of characters) of each entry title. Return a data frame that contains the title, author, and title length of each entry in descending order (longest title length at the top).

```python
newdf = df[['title','author','comments']]
countdf = []
for i in newdf['title']:
    countdf.append(len(i))
newdf['title_length'] = countdf
sort = newdf.sort_values(by='title_length',ascending=False)
sort
```
