### 7. Extract a list of entry titles.

```python
lst_entry_titles=[i.title for i in feedburner.entries]
lst_entry_titles
```

### 8. Calculate the percentage of "Four short links" entry titles.

```python
c=0
for i in lst_entry_titles:
    if i.find('Four short links') >= 0:
        c = c+1
        
Percentage = (c/len(lst_entry_titles))*100
print(Percentage,'%')
```

### 10. Count the number of entries per author and sort them in descending order.

```python
lst_authors=df_feeds.author.value_counts()
lst_row_authors=list(lst_authors.index)
lst_authors_num=lst_authors.tolist()

list_authors = [lst_row_authors,lst_authors_num]
dict_authors ={}
for i in range(len(lst_row_authors)):
    dict_authors.update({lst_row_authors[i]:lst_authors_num[i]})
dict_authors

from collections import OrderedDict
oda=(sorted(dict_authors.items(), key = lambda t: t[0]))
oda
```
