# Practicing english vocabulary. Part I (memorize)

```python
#Reading the text

with open("./texto.txt",'r') as f:
        content=f.read().lower().replace('.','').replace('-','').replace(',','').replace('  ',' ')

print(content)
```

```python
#Split the text into a words

splitting_text=content.split()
print(splitting_text)
```

```python
#Total words

len_splitting_text=len(splitting_text)
print(len_splitting_text)
print(splitting_text)
```

```python
#Droping small words, to obtain a better evaluation

small_words=[x for x in splitting_text if len(x) <=3]
print(len(small_words))
print(small_words)
for word in small_words:
    if word in splitting_text:
        splitting_text.remove(word)
print(splitting_text)
```

```python
#>>>>>>>>>>>>>>>>>> TO IMPROVE: TESTING UNIQUE WORDS
```

```python
#Total words after drop small words

len_splitting_text=len(splitting_text)
print(len_splitting_text)
```

```python
#User words
user_words=input("Write the words that you remember: ")
```

```python
#Splitting user words

#print(user_words)
split_user_words=user_words.split()
print(split_user_words)
print(len(split_user_words))
```

```python
#Evaluating words in original text

counter=0
for word in split_user_words:
    if word in splitting_text:
        print(word, 'True')
        counter +=1
    else:
        print(word, 'False')
print(counter)
```

```python
#Calculating user performance

percent=int((counter/len_splitting_text)*100)
if percent >= 50:
    print('Percentage of words remembered:',percent,'%','\nTerrific! You have good memory \(^_^)/')
elif percent <= 50:
    print('Percentage of words remembered:',percent,'%','\nUhhhmm You can do it better... Try again! T_T')
```

# Practicing english vocabulary. Part II (create)

```python
from math import factorial
```

```python
user_word=input("Write a word: ")
```

```python
print(user_word)
```

```python
n=(len(user_word))
print(n)
```

```python
r=n-1
print(r)
```

```python
p=int(factorial(n)/factorial(n-r))
print(p)
```

```python
print('There are',p, 'possible combinations of', r, 'letters, based on each letter of the word:',user_word)
```

```python
user_combinations=input("Please, write posible combinations : ")
```

```python
list_user_combination=user_combinations.split()
print(list_user_combination)
```

```python
#>>>>>>>>>>>>>>>>>> TO IMPROVE: EVALUATE IF THE WORDS EXISTS ON A REAL DICTIONARY
```

```python
brainly=int((len(list_user_combination)/p)*100)
print(brainly)
```

```python
print('You made', brainly,'% of the posible combinations :)')
```

```python
from gtts import gTTS

def tts(text_file, lang, name_file):
    with open(text_file, "r") as file:
        text = file.read()
    file = gTTS(text=text,lang=lang)
    filename = name_file
    file.save(filename)

tts("texto.txt", "ES", "texto.mp3")
```
