```python
# 1 Import libraries.
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# 2 Define the initial API endpoint URL.
url = 'https://rickandmortyapi.com/api/'
```

```python
# 3 Make request to obtain data of the 1st page. 
# . Flatten the data and store it in an empty object variable.
char = 'character/'
url_char = url + char
url_char_1 = url_char + '?page=1'

response = requests.get(url_char_1)
results = response.json()
#print(results)
print(url_char_1)
#https://rickandmortyapi.com/api/character/?page=1
```

```python
flattened_data = json_normalize(results)
#print(flattened_data['results'][0])
```

```python
flattened_data
```

```python
page_data = flattened_data
ch_data = json_normalize(page_data['results'][0])
ch_data.head()
```

```python
# 4 Find out the total page count in the 1st page data.
tot_pages = flattened_data['info.pages'][0]     
print('Tot. Pages: {}'.format(tot_pages))
```

```python
# 5 Use a FOR loop to make requests for the additional pages from 2 to page count.
# . Append the data of each additional page to the flatterned data object.
from IPython.display import clear_output

for page_num in range(1,int(tot_pages)):
    url_char_p = 'https://rickandmortyapi.com/api/character/?page={}'.format(page_num+1)
    response = requests.get(url_char_p)
    results = response.json()
    f_data = json_normalize(results)
    page_data=page_data.append(f_data)
    ch_data = ch_data.append(json_normalize(f_data['results'][0]))
    clear_output(wait=True)
    print('Loading..{}/{}'.format(page_num+1,tot_pages))
#flattened_data



#https://rickandmortyapi.com/api/character/?page=3
```

```python
ch_data = ch_data.set_index('id')
ch_data
```
