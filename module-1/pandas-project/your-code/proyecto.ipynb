# Drop Duplicates

```python
# elimina todas las filas que contienen valores iguales, que para este caso es nan en todas sus columnas
duplicados = shark_info.drop_duplicates().reset_index()
duplicados.sample(10)
```

# comparacion original vs drop duplicates

```python
print('tamaño de las filas en el archivo originial',shark_info.shape)
print('tamaño de las filas en el archivo que elimina valores duplicados',duplicados.shape)
```

```python
# reorganizacion de las columnas de la mas importante a menos importante 
cols = ['Case Number', 'Date', 'Year','Type','Country', 'Area','Location',
       'Activity', 'Name', 'Sex ', 'Age', 'Fatal (Y/N)','Species ','Investigator or Source',
       'original order','Case Number.1', 'Case Number.2','pdf', 'href formula', 'href','Unnamed: 22','Unnamed: 23']
shark_clean = pd.DataFrame(duplicados, columns=cols)
shark_clean = shark_clean
shark_clean
```

# primera columna 'Case Number'

```python
# limpiando primera columna de tabla
shark_clean['Case Number'] = shark_clean['Case Number'].str.replace('\.[a-z]+|\.[A-Z]+','', regex=True).fillna('2010.01.01')
shark_clean.dtypes
```

```python
shark_clean['Case Number'] = shark_clean['Case Number'].str.replace('xx','2010.01.01')
shark_clean
```

# limpieza de las fechas date 

```python
shark_clean['Date'] = shark_clean['Date'].str.replace('Reported| Early| Late|\.[a-z]|\.[A-Z]|Between May &' ,'').fillna('01-01-2000')
shark_clean.sample(20)
```

```python
shark_clean['Date'] = pd.to_datetime(shark_clean['Date'], errors='coerce', infer_datetime_format=True)
shark_clean.dtypes
```

```python
shark_clean.head()
```

# limpieza de year

```python
'''
convierte los valores de años a str
'''
shark_clean['Year']=shark_clean['Year'].fillna(shark_clean['Year'].mean())
shark_clean.head(10)
```

```python
shark_clean['Year']= shark_clean['Year'].astype({'Year':'str'})
shark_clean.head(10)
```

```python
shark_clean['Year'] = shark_clean['Year'].str.replace('\.[0-9]+','',regex=True)
shark_clean.head(10)
```

```python
shark_clean['Year'] = shark_clean['Year'].apply(pd.to_numeric)
shark_clean.sample(10)
```

```python
shark_clean['Year'] = shark_clean['Year'].astype({'Year':'int'})
shark_clean.dtypes
```

```python
aux_fun = lambda x : int(x)
shark_clean['Year'] = shark_clean['Year'].map(aux_fun)
shark_clean.head(10)
```

```python
shark_clean.dtypes
```

# limpieza de type

```python
shark_clean['Type']= shark_clean['Type'].str.replace('Invalid|0','Unknown',regex=True).fillna('Unknown')
shark_clean.sample(60)
```

```python
shark_clean.Type.unique
```

```python
def tipo(x):
    if 'Boat' in x:
        return 'Boating'
    elif 'Unprov' in x:
        return 'Unprovoked'
    elif 'Unknown' in x:
        return 'Unknown'
    elif 'Prov' in x:
        return 'Provoked'
    else:
        return 'Unknown'

shark_clean['Type'] = shark_clean['Type'].apply(tipo)
```

```python
shark_clean.Type
```

# limpieza de Fatal (Y/N)

```python
columnas = {shark_clean.columns[11] : 'Fatal'}
shark_clean = shark_clean.rename(columns=columnas)
shark_clean['Fatal']
```

```python
shark_clean['Fatal'] = shark_clean['Fatal'].fillna('Unknown')
shark_clean['Fatal'].unique()
```

```python
shark_clean['Fatal'].unique()
```

```python
def fatalidad(x):
    if 'N' in x:
        return 'No'
    elif 'Y' in x:
        return 'Yes'
    elif ' N' in x:
        return 'No'
    elif 'N ' in x:
        return 'No'
    elif 'y' in x:
        return 'Yes'
    else:
        return 'Unkown'
        
```

```python
shark_clean['Fatal'] = shark_clean['Fatal'].apply(fatalidad)
shark_clean['Fatal'].unique()
```

# limpieza de Species 

```python
shark_clean = shark_clean.astype({'Species ': 'str'})
shark_clean['Species '] = shark_clean['Species '].apply(lambda x: x.lower().lstrip().rstrip())

shark_clean['Species '].apply(lambda x: ''.join([i for i in x if i.isalpha()]).replace('shark', '')).value_counts()
```

```python
def spec_filter(x):
    if 'tig' in x:
        return 'Tiger'
    elif 'whi' in x:
        return 'White'
    elif 'bull':
        return 'bull'
    else:
        return 'Unknown'
```

```python
for i in shark_clean['Species '].unique():
    print(i)
```

```python
shark_clean['Species '] = shark_clean['Species '].apply(spec_filter)
shark_clean['Species '].unique()
```
