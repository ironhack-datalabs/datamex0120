```python
#Filtrado de valores nulos.
tiburones_2 = tiburones.T
indices_nulos = []
for x in tiburones.index:
    if tiburones_2[x].isnull().all():
        indices_nulos.append(x)
indices_nulos
```

```python
tiburones[tiburones['Case Number']=='0'] #VisualizaciÃ³n de los casos etiquetados con 0.
```

```python
# Dado que los casos 0 no contienen informaciÃ³n relevante, se procede a su eliminaciÃ³n.
indices_a_borrar = tiburones[tiburones['Case Number']=='0'].index
tiburones.drop(indices_a_borrar, inplace = True)
tiburones
```

```python
#Los registros con valores nulos en la columna 'Fatal (Y/N)' no ofrecen informaciÃ³n
#importante, y debido a esto es necesario borrarlos.
indices_a_borrar = tiburones[tiburones['Fatal (Y/N)'].isnull()].index
tiburones.drop(indices_a_borrar, inplace = True)
tiburones
```

```python
#Se revisa la manera en la que estÃ¡ contenida la informaciÃ³n en el campo 'Fatal (Y/N)'.
tiburones['Fatal (Y/N)'].value_counts()
```

```python
#Se corrigen los datos de la columna 'Fatal (Y/N)'.
tiburones['Fatal (Y/N)'] = tiburones['Fatal (Y/N)'].replace(' N|N ','N',regex=True)
tiburones['Fatal (Y/N)'] = tiburones['Fatal (Y/N)'].replace('M|2017','UNKNOWN',regex=True)
tiburones['Fatal (Y/N)'] = tiburones['Fatal (Y/N)'].replace('y','Y',regex=True)
tiburones['Fatal (Y/N)'].value_counts()
```

```python
#Se eliminan los nulos de la columna 'Location'.
tiburones['Location'] = tiburones['Location'].fillna('Unknown')
tiburones.Location.value_counts()
```

```python
#Se eliminan los nulos de la columna 'Area'.
tiburones['Area'] = tiburones['Area'].fillna('Unknown')
tiburones.Area.value_counts()
```

```python
tiburones['Year'].value_counts()
```

```python
tiburones = tiburones.astype({'Year':'str'})
tiburones.dtypes
```

```python
tiburones.Year = tiburones.Year.apply(lambda x:x.replace('.0',''))
tiburones['Year'].value_counts()
```

```python
tiburones['Activity'] = tiburones['Activity'].fillna('Swimming')
tiburones['Activity'].value_counts()
```

```python
tiburones['Species '].value_counts()
```

```python
tiburones = tiburones.astype({'Species ':'str'})
tiburones['Species '] = tiburones['Species '].apply(lambda x:x.lower().strip())
tiburones['Species '].apply(lambda x: ''.join([y for y in x if y.isalpha()]).replace('shark','')).value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace('(\W|^)White\s{0,3}shark(\W|$)','White',regex=True)
tiburones['Species '].value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace('(\W|^)Tiger\s{0,3}shark(\W|$)|Tiger 3m','Tiger',regex=True)
tiburones['Species '].value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace('(\W|^)Hammerhead\s{0,3}shark(\W|$)','Hammerhead',regex=True)
tiburones['Species '].value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace('(\W|^)Blue\s{0,3}shark(\W|$)','Blue',regex=True)
tiburones['Species '].value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace("Shark involvement prior to death was not confirmed|Invalid|Shark involvement not confirmed|Shark involvement prior to death unconfirmed|6' shark|4' shark|1.8 m [6'] shark|[4.5' to 5'] shark|>1.8 m [6'] shark|3 m to 4.3 m [10' to 14'] shark|3' to 3.5Unknown|Most likely, a small shark|1.5 m to 2 m shark|1.8 m [6'] shark",'Unknown',regex=True)
tiburones['Species '].value_counts()
```

```python
tiburones['Species '] = tiburones['Species '].replace('(\W|^)Bull\s{0,3}shark(\W|$)','Bull',regex=True)
tiburones['Species '].value_counts()
```
