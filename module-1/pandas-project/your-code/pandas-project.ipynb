```python
def dateFilter(text):
    date = re.findall('[0-9]{2}-[A-Z|a-z]{3}-[0-9]{4}', text)
    if len(date) > 0:
        return date[0]
    else:
        date = re.findall('[0-9]{2}-[0-9]{2}-[0-9]{4}', text)
        if len(date) > 0:
            return date[0]
        else:
            return '00-Ene-0000'
```

```python
def notDate(text):
    date = re.findall('[0-9]{2}-[A-Z|a-z]{3}-[0-9]{4}', text)
    if len(date) > 0:
        return False
    else:
        date = re.findall('[0-9]{2}-[0-9]{2}-[0-9]{4}', text)
        if len(date) > 0:
            return False
        else:
            return True
```
