```python
dataClean['Country'] = dataClean['Country'].astype('str')
dataClean['Country'] = dataClean['Country'].str.rstrip()
dataClean['Country'] = dataClean['Country'].str.lstrip()
```

```python
dataClean['Country'] = dataClean['Country'].str.replace('?', '')
dataClean['Country'] = dataClean['Country'].str.replace('-', '')
dataClean['Country'] = dataClean['Country'].str.replace('ST HELENA, British overseas territory', 'ST HELENA')
dataClean['Country'] = dataClean['Country'].str.replace('Fiji', 'FIJI')
dataClean['Country'] = dataClean['Country'].str.replace('Sierra Leone', 'SIERRA LEONE')
dataClean['Country'] = dataClean['Country'].str.replace('Coast of AFRICA', 'AFRICA')

 
```

```python
def clean(age):
    try:
        age = int(age)
    except ValueError:
        age = 0
    if (age > 0 and age <= 100):
        return age
    else: return np.nan
```

```python
dataClean['Age'] = dataClean['Age'].apply(clean)
dataClean['Age'].value_counts()
```

```python
def clean(x):
    if isinstance(x, str):
        x = x.lower()
        if 'blue' in x:
            return 'Blue Shark'
        if 'white' in x:
            return 'White Shark'
        elif 'tiger' in x:
            return 'Tiger Shark'
        elif 'blacktip' in x:
            return "Blacktip Shark"
        elif 'bull' in x:
            return 'Bull Shark'
        elif 'nurse' in x:
            return 'Nurse Shark'
        elif 'mako' in x:
            return 'Nurse Shark'
        elif 'Wobbegong' in x:
            return 'Wobbegong Shark'
    else: return 'UNKNOWN'
    
dataClean["Species"] = dataClean["Species"].apply(lambda x: clean(x))
```

```python
def clean(x):
    if isinstance(x, str):
        x = x.lower()
        if 'florida' in x:
            return "Florida"
        elif 'new south wales' in x:
            return "New South Wales"
        elif 'queensland' in x:
            return "Queensland"
        elif 'hawaii' in x:
            return "Hawaii"
        elif 'california' in x:
            return "California"
        elif 'kwaZulu-natal' in x:
            return "KwaZulu-Natal"
        elif 'western cape province' in x:
            return "Western Cape Province"
        elif 'western australia' in x:
            return "Western Australia"
    else: return "UNKNOWN"
       
dataClean["Area"] = dataClean["Area"].apply(lambda x: clean(x))
```
