#### 4. Import the users table 

```python
users_table = pd.read_sql_query('SELECT * FROM stats.users', engine)
users_table.head(3)
```

#### 5. Rename Id column to userId


```python
users_table.rename(columns={'Id':'userId'},inplace=True)
users_table.head(3)
```

#### 6. Import the posts table. 

```python
posts_table = pd.read_sql_query('SELECT * FROM stats.posts', engine)
posts_table.head(3)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
#https://www.geeksforgeeks.org/python-pandas-dataframe-rename/
posts_table.rename(columns={'Id':'postId','OwnerUserId':'userId'},inplace=True)
posts_table.head(3)
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
# note: df[['col1', 'col3']]
df_users=users_table[['userId','Reputation','Views','UpVotes','DownVotes']]
df_users.head(3)
```

```python
df_posts=posts_table[['postId','Score','userId','ViewCount','CommentCount']]
df_posts.head(3)
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#https://chrisalbon.com/python/data_wrangling/pandas_join_merge_dataframe/
df_merged=pd.merge(df_users, df_posts, on='userId')
df_merged.head(10)
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
df_merged.isnull().sum()
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain.

```python
#Aunque los datos faltantes son el 53% del total de la columna, no limpiaria la columna por completo, más bien rellenaría con cero para dejar normalizado el df y más adelante poder ocupar esa inforamción si es necesario.
df_merged['ViewCount']=df_merged['ViewCount'].fillna(0)
df_merged.head(10)
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
df_merged.ViewCount = df_merged.ViewCount.astype(int)
df_merged.dtypes
```
