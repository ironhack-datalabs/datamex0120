# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
# name = Babelgum
list(companies.find({"name": "Babelgum"},{'_id': 0, 'name':1}))


```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
#list(companies.find({"name": "Babelgum"}))
comp_5k = companies.find({"number_of_employees": {'$gt':5000}},{'_id':0,'name':1,'number_of_employees':1})
comp_5k_sorted = comp_5k.sort("number_of_employees",-1).limit(20)
list(comp_5k_sorted)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
comp_00_05=companies.find({'$and':[{'founded_year': {'$gte':2000}},{'founded_year': {'$lte':2005}}]},
                          {'_id':0,'name':1,'founded_year':1})
list(comp_00_05)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
x =companies.find({'$and':[{'ipo.valuation_amount':{'$gte':100000000}},
                           {'founded_year':{'$lte':2010}}]},
                  {'_id':0,'name':1,'ipo':1})

list(x)

```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
comp3 = companies.find({'$and':[
    {'number_of_employees':{'$lte':1000}},
    {'founded_year':{'$lte':2005}}
]},
                      {'_id':0,'name':1,'number_of_employees':1})


list(comp3.sort('number_of_employees',-1).limit(10))
```

### 6. All the companies that don't include the `partners` field.

```python
#db.inventory.find( { item : { $exists: false } } )


companies_nopart = companies.find( {'partners': {'$exists':'false'}} ,{'_id':0,'name':1})
list(companies_nopart)
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
comp_cat_null = companies.find({'category_code': None},{'_id':0,'name':1})
list(comp_cat_null)
#   { _id: 1, item: null },
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
comp4 = companies.find({'$and':[{'number_of_employees':{'$gte':100}},
                        {'number_of_employees':{'$lte':1000}}]},
               {'_id':0,'name':1,'number_of_employees':1})
list(comp4)
```

### 9. Order all the companies by their IPO price descendently.

```python
 x =companies.find({'$and':[{'ipo.valuation_amount':{'$gte':100000000}},
                           {'founded_year':{'$lte':2010}}]},
                  {'_id':0,'name':1,'ipo.valuation_amount':1}).sort('ipo.valuation_amount',-1)
list(x)
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
comp6 = companies.find({'founded_month':{'$gte':7}},{'_id':0,'name':1}).sort('founded_month',-1).limit(1000)
list(comp6)
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
comp6 = companies.find({'$and':[{'$where':"this.deadpooled_year - this.founded_year >= 3"},{'deadpooled_year':{'$ne':None}},{'founded_year':{'$ne':None}},{'deadpooled_year':{'$gt':3}}]},
                            
                   {'_id':0,'name':1,'founded_year':1,'deadpooled_year':1}
    )

list(comp6)
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
comp9 = companies.find({'founded_year':{'$ne':None}},{'_id':0,'name':1,'founded_year':1}).sort('founded_year',1).limit(20)
list(comp9)
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
comp10 = companies.find({'$and':[{'founded_month':1},
                                {'founded_day':{'$gte':1}},
                                {'founded_day':{'$lte':7}}]},{'_id':0,'name':1}).sort('acquisition.price_amount',-1).limit(10)
list(comp10)
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
comp11 = companies.find({'$and':[{'category_code':'web'},{'number_of_employees':{'$gt':4000}}]},{'_id':0,'name':1,'number_of_employees':1}).sort('number_of_employees',-1)
list(comp11)
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.


```python
comp12 = companies.find({'$and':[{'acquisition.price_amount':{'$gt':100}},{'acquisition.price_currency_code':'EUR'}]},{'_id':0,'name':1})
list(comp12)
```

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
comp13 = companies.find(
                        {'$and':[{'acquisition.acquired_month':{'$gte':7}},
                                 {'acquisition.acquired_month':{'$lte':12}}]},
                        {'_id':0,'name':1,'acquisition':1},
                       )
list(comp13)
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.
