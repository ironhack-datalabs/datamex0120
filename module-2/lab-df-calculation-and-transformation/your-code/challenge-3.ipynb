We break down the main problem into several sub problems:

```python
# your code here
#orders_spent_Cou = orders.groupby(['CustomerID','Country'])[['amount_spent']].sum()
orders_spent_CiD = orders.groupby(['CustomerID'])[['amount_spent']].sum()
q_95 = orders_spent_CiD.quantile(.95)[0]
q_75 = orders_spent_CiD.quantile(.75)[0]

#orders_spent_CiD.head
print(q_95)
print(q_75)
```

```python
orders_spent_vip = orders_spent_CiD[orders_spent_CiD['amount_spent']>=q_95]
orders_spent_vip.sort_values(['amount_spent'], ascending=False, inplace=True)
orders_spent_vip['Type']='VIP'
orders_spent_vip.head()
```

```python
orders_spent_Pref = orders_spent_CiD[(orders_spent_CiD['amount_spent']<q_95)&(orders_spent_CiD['amount_spent']>=q_75)]
orders_spent_Pref.sort_values(['amount_spent'], ascending=False, inplace=True)
orders_spent_Pref['Type']='Preferred'
orders_spent_Pref.head()
```

```python
print(orders_spent_vip.shape)
print(orders_spent_Pref.shape)
```

```python
orders_spent_Ord = orders_spent_CiD[(orders_spent_CiD['amount_spent']<q_75)]
orders_spent_Ord.sort_values(['amount_spent'], ascending=False, inplace=True)
orders_spent_Ord['Type']='Ord'
orders_spent_Ord.head()
```

```python
def cat_cust(amount_spent):
    if amount_spent>=q_95:
        return 'VIP'
    elif amount_spent<q_95 and amount_spent>q_75:
        return 'Preferred'
    else:
        return 'Ordinary'
```

```python
orders_spent_CiD['Type'] = orders_spent_CiD['amount_spent'].apply(cat_cust)
```
