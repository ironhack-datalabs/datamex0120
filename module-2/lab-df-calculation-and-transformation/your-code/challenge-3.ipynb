```python
customers=pd.DataFrame(orders.groupby('CustomerID').sum()['amount_spent'])
customers
```

```python
#For part 2:...
q75=np.percentile(customers, 75)
q95=np.percentile(customers, 95)
VIP=customers[customers['amount_spent']>q95]
preferred=customers[(customers['amount_spent']<q95) & (customers['amount_spent']>q75)]
```

```python
#For part 3:...
customers['VIP/Preferred'] = np.where(customers['amount_spent']>q75, 'Preferred', 'Regular')
customers['VIP/Preferred'] = np.where(customers['amount_spent']>q95, 'VIP', customers['VIP/Preferred'])
customers
```
