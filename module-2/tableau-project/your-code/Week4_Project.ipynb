### Getting E-comm companies

```python
# Getting all ecommerce companies

ecom_query = { "category_code": "ecommerce" }
ecom_data = companies.find(ecom_query, {'_id':0,'name':1,'category_code':1, 'number_of_employees':1,'offices':1})
#len ecom_data = 688
#list(ecom_data)
ecom_data[0]['offices'][0]
```

```python
# Creating database from ecom_data

res = []

for elem in ecom_data:
    na = elem['name']
    num_emp = elem['number_of_employees']
    for office in elem['offices']:
        lat = office['latitude']
        lon = office['longitude']
        office = {'name':na, 'number_of_employees': num_emp, 'latitude':lat, 'longitude':lon}
        res.append(office)
```

```python
res
```

```python
res_df = pd.DataFrame(res)
res_df = res_df.sort_values(by='number_of_employees', ascending=False)
res_df = res_df.dropna(subset=['longitude'])
res_df = res_df.dropna(subset=['latitude'])
res_df = res_df.reset_index(drop=True)
res_df.head()
```

```python
#res_df.number_of_employees.value_counts()

res_df['latitude'] = res_df['latitude'].astype(float)
res_df['longitude'] = res_df['longitude'].astype(float)
```

```python
ecomm_coor = res_df[['latitude','longitude']]
heat_data = [[row['latitude'],row['longitude']] for index, row in ecomm_coor.iterrows()]
```

```python
map_comp = folium.Map(location=[39.381266, -97.922211],
                    zoom_start = 4)
HeatMap(heat_data).add_to(map_comp)
```

### Getting legal & finance companies

```python
# Getting legal and finance companies

legal_query = { "category_code": "legal"}
legal_data = companies.find(legal_query, {'_id':0,'name':1,'category_code':1, 'number_of_employees':1,'offices':1,'deadpooled_year':1})
legal_data[0]
```

```python
res2 = []

for elem in legal_data:
    na = elem['name']
    num_emp = elem['number_of_employees']
    for office in elem['offices']:
        lat = office['latitude']
        lon = office['longitude']
        office = {'name':na, 'number_of_employees': num_emp, 'latitude':lat, 'longitude':lon}
        res2.append(office)
```

```python
leg_df = pd.DataFrame(res2)
leg_df = leg_df.sort_values(by='number_of_employees', ascending=False)
leg_df = leg_df.dropna(subset=['longitude'])
leg_df = leg_df.dropna(subset=['latitude'])
leg_df = leg_df.reset_index(drop=True)
leg_df.head()
```

```python
finance_query = { "category_code": "finance"}
finance_data = companies.find(finance_query, {'_id':0,'name':1,'category_code':1, 'number_of_employees':1,'offices':1,'deadpooled_year':1})
finance_data[0]
```

```python
res3 = []

for elem in finance_data:
    na = elem['name']
    num_emp = elem['number_of_employees']
    for office in elem['offices']:
        lat = office['latitude']
        lon = office['longitude']
        office = {'name':na, 'number_of_employees': num_emp, 'latitude':lat, 'longitude':lon}
        res3.append(office)
```

```python
fin_df = pd.DataFrame(res3)
fin_df = fin_df.sort_values(by='number_of_employees', ascending=False)
fin_df = fin_df.dropna(subset=['longitude'])
fin_df = fin_df.dropna(subset=['latitude'])
fin_df = fin_df.reset_index(drop=True)
fin_df.head()
```

```python
coor_leg = []
for index, row in leg_df.iterrows():
    coor_leg.append([row['name'],row['latitude'], row['longitude']])
    
coor_fin = []
for index, row in fin_df.iterrows():
    coor_fin.append([row['name'],row['latitude'], row['longitude']])
```

```python
print(len(coor_leg))
print(len(coor_fin))

for elem in coor_leg:
    folium.CircleMarker([elem[1], elem[2]], radius=10, popup=elem[0],
                        color='#3186cc', fill=True, fill_color='#3186cc').add_to(map_comp)

    
    
for elem in coor_fin:
    folium.Marker([elem[1], elem[2]], radius=5, popup=elem[0], icon=folium.Icon(color='red')).add_to(map_comp)
    
```

### Getting Companies that were on higher than the 75percentile in acquisition price amount

```python
acq_query = {'acquisition.price_amount':{'$gte':0}}
acq_data = companies.find(acq_query, {'_id':0,'name':1,'acquisition.price_amount':1, 'acquisition.acquired_year':1,
                                      'offices':1})

#acq_data[0]

res4=[]

for elem in acq_data:
    na = elem['name']
    acq_price = elem['acquisition']['price_amount']
    acq_year = elem['acquisition']['acquired_year']
    
    for office in elem['offices']:
        lat = office['latitude']
        lon = office['longitude']
        office = {'name':na, 'acq_price':acq_price, 'acq_year':acq_year, 'latitude':lat, 'longitude':lon}
        res4.append(office)
```

```python
acq_df = pd.DataFrame(res4)

acq_df = acq_df.sort_values(by='acq_year', ascending=False)
acq_df = acq_df.dropna(subset=['longitude'])
acq_df = acq_df.dropna(subset=['latitude'])
acq_df = acq_df.dropna(subset=['acq_year'])
acq_df = acq_df[(acq_df.acq_year>=2010)&(acq_df.acq_price>=acq_df.acq_price.quantile(.75))]
acq_df = acq_df.reset_index(drop=True)
acq_df.head()
```

```python
coor_acq = []
for index, row in acq_df.iterrows():
    coor_acq.append([row['name'],row['latitude'], row['longitude']])
    
print(len(coor_acq))

for elem in coor_acq:
    folium.Marker([elem[1], elem[2]], radius=8, popup=elem[0], icon=folium.Icon(color = 'green') ).add_to(map_comp)
```
