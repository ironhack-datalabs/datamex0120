```python
# El cliente busca lo siguiente:

# - Establecer empresa de juegos de video.

# - Cercano a empresas de juegos de video, software, hardware, mobile, network hosting y web.

# - Empresas fundadas en los últimos 30 años.

# - De preferencia cerca de empresas de origen californiano.
```

```python
filtro_empresas = pd.DataFrame(db.find({'deadpooled': None, '$or': [{'category_code':'software'},
                                          {'category_code':'hardware'}, 
                                          {'category_code':'mobile'}, 
                                          {'category_code':'network_hosting'},
                                          {'category_code':'web'}],
                                  'founded_year':{'$gt':1990},
                                  'offices.0.state_code':'CA'},
                                {'_id':0, 'name':1, 'category_code':1, 'founded_year':1, 'offices':1}))
filtro_empresas.shape
```

```python
filtro_empresas.head()
```

```python
filtro_empresas['offices'] = filtro_empresas['offices'].apply(lambda x : x[0])
```

```python
datos_oficinas = json_normalize(filtro_empresas.offices)
```

```python
df_empresas = pd.concat([filtro_empresas[['name', 'founded_year', 'category_code']], datos_oficinas], axis=1)
df_empresas.head()
```

```python
df_empresas.shape
```

```python
df_empresas.drop(df_empresas[df_empresas['latitude'].isnull()].index, inplace=True)
df_empresas.drop_duplicates(inplace=True)
```

```python
df_empresas.shape
```

```python
df_empresas.head()
```

```python
df_empresas.reset_index(inplace=True)
```

```python
mapa = folium.Map(location = [40, -100], tiles = 'openstreetmap', zoom_start = 4.4)
for indice in df_empresas.index:
    lugar = df_empresas.loc[indice]
    texto_emergente = '<i>' + lugar['name'] + '</i>'
    folium.Marker([lugar.latitude, lugar.longitude], popup = texto_emergente, tooltip = lugar.category_code).add_to(mapa)
mapa
```

```python
# Tomando en cuenta lo observado en el mapa, el cliente ha optado por establecerse en un
# lugar céntrico del estado de California.
```

```python
latitud = df_empresas.latitude.mean()
longitud = df_empresas.longitude.mean()
mapa = folium.Map(location = [40, -100], tiles = 'openstreetmap', zoom_start = 4.4)
folium.Circle([latitud, longitud], radius = 180000, icon=folium.Icon()).add_to(mapa)
mapa
```
